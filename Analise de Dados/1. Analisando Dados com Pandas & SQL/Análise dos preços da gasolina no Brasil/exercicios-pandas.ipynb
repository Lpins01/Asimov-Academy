## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

### 1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
df_2000 = pd.read_csv("gasolina_2000+.csv", index_col=0)
df_2010 = pd.read_csv("gasolina_2010+.csv", index_col=0)

df = pd.concat([df_2000, df_2010])
```

### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
df["DATA INICIAL"] = pd.to_datetime(df["DATA INICIAL"])
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
df["ANO-MÊS"] = df["DATA FINAL"].apply(lambda x: "{}".format(x.year)) + df["DATA FINAL"].apply(lambda x: "-{:02d}".format(x.month))
```

### 6. Utilizando o value_counts(), liste todos os tipos de produtos contidos na base de dados.

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
df_GasolinaComum = df[df["PRODUTO"] == "GASOLINA COMUM"]
```

### 11. Qual a média de preço dos estados da região sul em 2012?

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
df_GasolinaComum["ANO"] = df_GasolinaComum["DATA FINAL"].dt.year
df_Rio = df_GasolinaComum[df_GasolinaComum["ESTADO"] == "RIO DE JANEIRO"][["ANO", "PREÇO MÉDIO REVENDA"]].groupby("ANO").last()
```

```python
df_Rio["VARIAÇÃO PERCENTUAL"] = ((df_Rio - df_Rio.shift(1))/ df_Rio.shift(1)) * 100
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
df_Max = df_GasolinaComum.groupby("ANO-MÊS").max()["PREÇO MÉDIO REVENDA"]
df_Min = df_GasolinaComum.groupby("ANO-MÊS").min()["PREÇO MÉDIO REVENDA"]

idx_max = df_GasolinaComum.groupby("ANO-MÊS")["PREÇO MÉDIO REVENDA"].idxmax()
idx_min = df_GasolinaComum.groupby("ANO-MÊS")["PREÇO MÉDIO REVENDA"].idxmin()
```

```python
df_Diff = pd.DataFrame()
```

```python
df_Diff["ABS_DIFF"] = df_Max - df_Min
df_Diff["PERCENT_DIFF"] = (df_Max - df_Min) / df_Min * 100
df_Diff["MAX"] = df_Max
df_Diff["MIN"] = df_Min
```

```python
'''
df_GasolinaComum.loc[idx_max]["ESTADO"] gera um dataframe onde os indices sao os indices originais, como no df_Diff os indices sao datas, podemos usar o argumento
.values para pegar somente os valores do dataframe

IMPORTANTE: .values so funciona nesse caso porque os dataframes que queremos juntar possuem o mesmo tamanho
'''

df_Diff["ESTADO_MAX"] = df_GasolinaComum.loc[idx_max]["ESTADO"].values
df_Diff["ESTADO_MIN"] = df_GasolinaComum.loc[idx_min]["ESTADO"].values
```
